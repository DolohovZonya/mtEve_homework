```python
from Bio import AlignIO

alignment = AlignIO.read("all_humans_mafft_alignment.fasta", "fasta")
root = alignment[-1].seq #the most ancient taxon stands in the end of the file
snp_max = 0
idshnik = ''
for record in alignment:
  snp_count = 0
  for i, base in enumerate(record.seq):
        if base != root[i]:
            snp_count += 1
        if snp_count > snp_max:
          snp_max = snp_count
          idshnik = record.id
        

print(f"max_distance: {snp_max, idshnik}")
```
